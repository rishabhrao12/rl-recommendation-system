# Training Loop

```python
import numpy as np
import pandas as pd
import os
from sklearn.metrics.pairwise import cosine_similarity
import json
import random
```

```python
# Load data
df_subset = pd.read_csv("news_subset_2k.csv")
embeddings = np.load("embeddings_2k.npy")

n_articles = len(df_subset)
embedding_dim = embeddings.shape[1]

# Normalize embeddings for cosine similarity
embeddings = embeddings / np.linalg.norm(embeddings, axis=1, keepdims=True)

# Load model state if it exists
if os.path.exists("q_table.npy"):
    q_table = np.load("q_table.npy")
    user_pref_vector = np.load("user_pref_vector.npy")
    with open("user_weight.json", "r") as f:
        total_weight = json.load(f)["total_weight"]
else:
    q_table = np.zeros((n_articles, 4))
    user_pref_vector = np.zeros(embedding_dim)
    total_weight = 1e-5

# Config
action_map = {'like': 0, 'share': 1, 'read': 2, 'not_interested': 3}
reward_map = {'like': 1.0, 'share': 1.5, 'read': 0.5, 'not_interested': -1.0}
epsilon = 0.1  # exploration rate

# State tracking
seen_articles = set()
interaction_log = []
```

```python
# Helper functions
def update_user_pref(embedding, action):
    global user_pref_vector, total_weight
    reward = reward_map[action]
    if reward > 0:
        user_pref_vector = (user_pref_vector * total_weight + reward * embedding) / (total_weight + reward)
        total_weight += reward
    elif reward < 0:
        user_pref_vector = (user_pref_vector * total_weight + reward * embedding) / max(total_weight + abs(reward), 1e-5)
        total_weight += abs(reward)

def score_article(idx):
    sim = cosine_similarity(user_pref_vector.reshape(1, -1), embeddings[idx].reshape(1, -1))[0][0]
    q_score = np.max(q_table[idx])
    return 0.7 * q_score + 0.3 * sim

def recommend_article():
    candidates = [i for i in range(n_articles) if i not in seen_articles]
    if not candidates:
        return None  # all articles seen

    if random.random() < epsilon:
        return random.choice(candidates)  # explore
    return max(candidates, key=score_article)  # exploit
```

```python
# 🔁 Interactive training loop
for step in range(20):  # increase as needed
    article_idx = recommend_article()
    if article_idx is None:
        print("\n✅ All articles seen! Training complete.")
        break

    article = df_subset.iloc[article_idx]
    seen_articles.add(article_idx)

    print(f"\n📢 Recommended Article #{article_idx}")
    print(f"📰 Title: {article['title']}")
    print(f"📂 Category: {article['category']}")

    action = input("Your action? (like/share/read/not_interested): ").strip().lower()
    while action not in action_map:
        action = input("Invalid. Choose from like, share, read, not_interested: ").strip().lower()

    reward = reward_map[action]
    update_user_pref(embeddings[article_idx], action)

    action_idx = action_map[action]
    q_table[article_idx, action_idx] += 0.1 * (reward - q_table[article_idx, action_idx])  # learning rate = 0.1

    interaction_log.append((article['title'], article['category'], action, reward))
```

```python
# Save state
np.save("q_table.npy", q_table)
np.save("user_pref_vector.npy", user_pref_vector)
with open("user_weight.json", "w") as f:
    json.dump({"total_weight": total_weight}, f)

# Save log
log_df = pd.DataFrame(interaction_log, columns=["Title", "Category", "Action", "Reward"])
log_df.to_csv("interaction_log.csv", index=False)
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def create_interactive_user_insights(log_path="interaction_log.csv"):
    # Load interaction log
    log_df = pd.read_csv(log_path)

    # Prepare data
    action_counts = log_df['Action'].value_counts()
    category_rewards = log_df.groupby('Category')['Reward'].sum().sort_values(ascending=False)
    liked_df = log_df[log_df['Action'] == 'like']
    liked_counts = liked_df['Category'].value_counts()
    log_df['Cumulative Reward'] = log_df['Reward'].cumsum()
    top_liked = liked_df['Title'].value_counts().head(5)

    # Create subplot layout
    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=[
            "User Actions Distribution",
            "Total Reward per Category",
            "Articles Liked per Category",
            "Cumulative Reward Over Time",
            ""
        ],
        specs=[[{"type": "bar"}, {"type": "bar"}],
               [{"type": "bar"}, {"type": "scatter"}]])

    # Plot 1: Action distribution
    fig.add_trace(go.Bar(x=action_counts.index, y=action_counts.values, name="Actions", marker_color='skyblue'), row=1, col=1)

    # Plot 2: Rewards by category
    fig.add_trace(go.Bar(x=category_rewards.index, y=category_rewards.values, name="Rewards", marker_color='lightgreen'), row=1, col=2)

    # Plot 3: Likes by category
    fig.add_trace(go.Bar(x=liked_counts.index, y=liked_counts.values, name="Likes", marker_color='coral'), row=2, col=1)

    # Plot 4: Cumulative reward over time
    fig.add_trace(go.Scatter(y=log_df['Cumulative Reward'], mode='lines+markers', name='Cumulative Reward'), row=2, col=2)

    ## Plot 5: Top liked articles
    #fig.add_trace(go.Bar(y=top_liked.index[::-1], x=top_liked.values[::-1], orientation='h', name="Top Liked", marker_color='orchid'), row=3, col=1)

    # Layout styling
    fig.update_layout(height=900, width=1000, title_text="📊 User Recommendation Insights", showlegend=False)
    return fig
```

```python
fig = create_interactive_user_insights()
fig.show()
```
